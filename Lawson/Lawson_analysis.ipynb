```python
# Load data
import numpy as np
X = np.loadtxt('expressions.txt')
genes = np.loadtxt('genes.txt',dtype=str)
```

```python
# Functions for performing PCA, t-SNE

import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn import manifold
from sklearn import cluster

# obtain PCs and plot by batch number
def jz_PCA(X):
    X_zm = X
    X_bar = np.mean(X,axis=0)
    U, s, V = np.linalg.svd(np.dot(X_zm,np.transpose(X_zm)), full_matrices=True)
    return U

# obtain two major directions from t-SNE and plot by batch number
def jz_tSNE(X):
    tsne = manifold.TSNE(n_components=2, random_state=0);
    X_tsne = tsne.fit_transform(X);
    return X_tsne
```

```python
# PCA 
%matplotlib inline

U = jz_PCA(np.transpose(X))
plt.figure(figsize=(10,10))
plt.scatter(U[:,0],U[:,1])
plt.show()
```

```python
# t-SNE 
%matplotlib inline

U = jz_tSNE(np.transpose(X))
plt.figure(figsize=(10,10))
plt.scatter(U[:,0],U[:,1])
plt.show()
```
